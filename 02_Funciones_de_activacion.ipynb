# **Funciones en Deep Learning**


## **Librerías**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## **Funciones de activación**

### **1. Función Sigmoide**

```python
def sigmoide(x):
  return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(10, -10, 100)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

fig.patch.set_facecolor('#21252b') 
ax.set_facecolor('#282c34') 

plt.plot(x, sigmoide(x))
plt.title('Función sigmoide', color='white')

plt.xticks(color='white', size=12)
plt.yticks(color='white', size=12)

plt.show()
```

### **2. Función signo**

```python
def signo(x):
  return np.piecewise(x, [x<0.0, x>0.0], [-1, 1])
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

fig.patch.set_facecolor('#21252b') 
ax.set_facecolor('#282c34') 

plt.plot(x, signo(x))
plt.title('Función signo', color='white')

plt.xticks(color='white', size=12)
plt.yticks(color='white', size=12)

plt.show()
```

### **3. Función escalonada**

```python
def escalonada(x):
  return np.piecewise(x, [x<0.0, x>0.0], [0, 1])
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

fig.patch.set_facecolor('#21252b') 
ax.set_facecolor('#282c34') 

plt.plot(x, escalonada(x))
plt.title('Función escalonada', color='white')

plt.xticks(color='white', size=12)
plt.yticks(color='white', size=12)

plt.show()
```

### **4. Función tangente hiperbólica**

```python
fig, ax = plt.subplots(figsize=(10, 5))

fig.patch.set_facecolor('#21252b') 
ax.set_facecolor('#282c34') 

plt.plot(x, np.tanh(x))
plt.title('Función tangete hiperbólica', color='white')

plt.xticks(color='white', size=12)
plt.yticks(color='white', size=12)

plt.show()
```

### **5. Función ReLu**

```python
def relu(x):
  if x <= 0:
    return 0
  else:
    return x
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

fig.patch.set_facecolor('#21252b') 
ax.set_facecolor('#282c34') 

plt.plot(x, [relu(i) for i in x])
plt.title('Función ReLu', color='white')

plt.xticks(color='white', size=12)
plt.yticks(color='white', size=12)

plt.show()
```

## **Función de Perdida**

### **1. Error Cuadratico Medio**

```python
def mse(y_pred, y_real, derivate=False):
  if derivate==True:
      return (y_pred - y_real)
  else:
    return np.mean((y_pred - y_real)**2)
```

```python
valor_real = np.array([0, 0, 1, 1])
prediccion = np.array([0.9, 0.5, 0.2, 0.12])
```

```python
mse(prediccion, valor_real)
```

```python
mse(prediccion, valor_real, derivate=True)
```
